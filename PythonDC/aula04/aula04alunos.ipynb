#### COMEÇANDO O DESAFIO



Analise os dados com base nas perguntas.

1 - Temos dados de quantas escolas?

```python
total_escola = len(df_escola_estudantes["Nome_Escola"].unique())
print(f"O total de escolas únicas é: {total_escola} ")
```

2 - Qual o total de registros de estudantes na base de dados?


```python
total_estudantes = df_escola_estudantes["ID_Estudante"].count() #contando a quantidade de estudantes utilizando a coluna "ID_Estudante" do dataframe df_escola_estudantes
print(f"O total de estudantes é: {total_estudantes} ")
```

3 - Qual o orçamento total considerando todas as escolas?

```python
#primeira solução errada
orcamento_total = (df_escola_estudantes["Orcamento_Anual"].sum()) #calculando o orçamento total utilizando a coluna "Orçamento" do dataframe df_escola_estudantes
print(f"O orçamento total é: {orcamento_total} ")
```

```python
##formatação correta do orçamento total utilizando a biblioteca locale para formatar o número como moeda brasileira

import locale

locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

```python
import locale

locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

# Segunda SOLUÇÃO (CORRETA)
total_orcamento_unico = (
    df_escola_estudantes[['Nome_Escola', 'Orcamento_Anual']]
    .drop_duplicates()
)

total_orcamento_unico = total_orcamento_unico["Orcamento_Anual"].sum()

print(f'💰 A soma total do orçamento anual das escolas é: {locale.currency(total_orcamento_unico, grouping=True)}')

```

4 - Qual a média da nota dos alunos em Redação?

```python
media_redacao = df_escola_estudantes["Nota_Redacao"].mean()

print(f'A média de Redação é: {media_redacao:.2f}')
```

5 - Qual a média da nota dos alunos em Matemática?


```python
media_matematica = df_escola_estudantes["Nota_Matematica"].mean()

print(f'A média de Matemática é: {media_matematica:.2f}')
```

6 - Considerando que a nota de aprovação é 70, quantos alunos foram aprovados em Redação? (Entegue o resultado em valor absoluto e percentual).

```python
#Valor Absoluto

#Com o count precisa passar a coluna Nome Estudante

cont_estud_aprov_red = df_escola_estudantes[df_escola_estudantes["Nota_Redacao"] >= 70].count()["Nome_Estudante"]

print(f'📝 Total de estudantes aprovados em redação: {cont_estud_aprov_red}')
```

```python
# Com o shape é já faço a contagem da dimensão, não preciso passar a coluna de estudante
cont_estud_aprov_red = df_escola_estudantes[df_escola_estudantes["Nota_Redacao"] >= 70].shape[0]

print(f'📝 Total de estudantes aprovados em redação: {cont_estud_aprov_red}')
```

```python
##Percentual
percentual_aprov_redacao = cont_estud_aprov_red / float(total_estudantes)*100
print(f'📝 Percentual de Aprovação em Redação é: {percentual_aprov_redacao:.2f}')
```

7 - Considerando que a nota de aprovação é 70, quantos alunos foram aprovados em Matemática? (Entregue o resultado em valor absoluto e percentual)


```python
# Com o shape é já faço a contagem da dimensão, não preciso passar a coluna de estudante
cont_estud_aprov_mat = df_escola_estudantes[df_escola_estudantes["Nota_Matematica"] >= 70].shape[0]

print(f'📝 Total de estudantes aprovados em matemática: {cont_estud_aprov_mat}')
```

```python
##Percentual

percentual_aprov_matematica = cont_estud_aprov_mat / float(total_estudantes)*100
print(f"📝 Percentual de Aprovação em Matemática é: {percentual_aprov_matematica:.2f}")
```

8 - Considerando que a noa de aprovação é 70, quantos alunos foram aprovados em Matemática e Redação?
(Entregue o resultado em valor absoluto e percentual)

```python
cont_aprov_total = df_escola_estudantes[(df_escola_estudantes["Nota_Matematica"] >= 70)
                                  & (df_escola_estudantes["Nota_Redacao"] >= 70)].shape[0]
print(f'📝 Total de estudantes aprovados em matemática e redação: {cont_aprov_total}')
```

```python
##Percentual

percentual_aprov_total = cont_aprov_total / float(total_estudantes)*100
print(f'📝 Percentual de Aprovação em Matemática e Redação é: {percentual_aprov_total:.2f}')

```

9 - Crie um dataframe com os resultados das questões de 1 a 8 que você calculou acima.

10 - Formate as colunas "Total de Estudantes" e "Total Orçamento" ajustando as casas decimais.

```python

##SIMPLES
df_performance_escolar["Total de Estudantes"] = df_performance_escolar["Total de Estudantes"].map("{:,}".format)
df_performance_escolar["Total Orçamento"] = df_performance_escolar["Total Orçamento"].map("${:,.2f}".format)
```

```python
##COMPLEXA
def fmt_br(valor, prefixo="", sufixo="", decimais=2):
    """Formata número no padrão brasileiro: 1.234,56"""
    valor = float(str(valor).replace(",", "").replace("$", "").strip())  # garante que é número
    formatado = f"{valor:,.{decimais}f}"
    formatado = formatado.replace(",", "|").replace(".", ",").replace("|", ".")
    return f"{prefixo}{formatado}{sufixo}"

df_performance_escolar["Total de Escolas"]        = df_performance_escolar["Total de Escolas"].map(lambda x: fmt_br(x, decimais=0))
df_performance_escolar["Total de Estudantes"]     = df_performance_escolar["Total de Estudantes"].map(lambda x: fmt_br(x, decimais=0))
df_performance_escolar["Total Orçamento"]         = df_performance_escolar["Total Orçamento"].map(lambda x: fmt_br(x, prefixo="R$ "))
df_performance_escolar["Nota Média em Redação"]   = df_performance_escolar["Nota Média em Redação"].map(lambda x: fmt_br(x))
df_performance_escolar["Nota Média em Matemática"]= df_performance_escolar["Nota Média em Matemática"].map(lambda x: fmt_br(x))
df_performance_escolar["% Aprovação Redação"]     = df_performance_escolar["% Aprovação Redação"].map(lambda x: fmt_br(x, sufixo="%"))
df_performance_escolar["% Aprovação Matemática"]  = df_performance_escolar["% Aprovação Matemática"].map(lambda x: fmt_br(x, sufixo="%"))
df_performance_escolar["% Aprovação Total"]       = df_performance_escolar["% Aprovação Total"].map(lambda x: fmt_br(x, sufixo="%"))

df_performance_escolar
```
