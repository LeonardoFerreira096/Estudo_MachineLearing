# 📄 PARTE 1 — Arquivos TXT

Arquivos TXT são texto puro — sem formatação especial. São ótimos para:
- ✅ Logs de sistema (registros de eventos)
- ✅ Relatórios simples gerados automaticamente  
- ✅ Listas e anotações

**🧠 Analogia:** Um arquivo TXT é como um bloco de notas digital. Simples, universal e confiável.

### Modos de abertura (`open`)

| Modo | Significado | O que faz |
|---|---|---|
| `"r"` | read | Lê o arquivo (erro se não existir) |
| `"w"` | write | Cria ou **sobrescreve** o arquivo |
| `"a"` | append | **Acrescenta** ao final do arquivo |


### ✏️ 1.1 — Criando e escrevendo um arquivo TXT

Vamos simular um analista que precisa registrar as vendas do dia num arquivo de texto.

````python
# Dados que queremos salvar (simula dados de vendas do dia)
conteudo = """=== RELATÓRIO DE VENDAS - 15/01/2025 ===

Analista de Dados é a profissão que mais tem crescido em todo mundo.
Esses profissionais precisam se especializar em Programação, Estatística e Machine Learning.
E claro, em Big Data.
"""

print("Conteúdo que vamos salvar:")
print("-" * 40)
print(conteudo)
````

````python
# ✅ FORMA RECOMENDADA: usando "with open"
# O arquivo é fechado AUTOMATICAMENTE ao sair do bloco with
# Sempre use encoding='utf-8' para suportar acentos!

with open('arquivos/relatorio.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.write(conteudo)

print('✅ Arquivo "relatorio.txt" criado com sucesso!')
````

### 📖 1.2 — Lendo um arquivo TXT

Existem três formas principais de ler um arquivo TXT:

````python
# Método 1: read() — lê o arquivo INTEIRO de uma vez
# Use quando o arquivo é pequeno e você quer tudo de uma vez

with open('arquivos/relatorio.txt', 'r', encoding='utf-8') as arquivo:
    texto_completo = arquivo.read()

print("📄 Conteúdo completo:")
print(texto_completo)
print(f"Total de caracteres: {len(texto_completo)}")
````

````python
# Método 2: readlines() — lê todas as linhas e retorna uma LISTA
# Use quando precisa processar cada linha separadamente

with open('arquivos/relatorio.txt', 'r', encoding='utf-8') as arquivo:
    linhas = arquivo.readlines()

print(f"Total de linhas: {len(linhas)}")
print("\nCada linha:")
for i, linha in enumerate(linhas, 1):
    print(f"  Linha {i}: {repr(linha)}")  # repr() mostra o \n
````

````python
# Método 3: loop direto — mais eficiente para arquivos GRANDES
# Python processa uma linha por vez, economizando memória

print("Linhas não-vazias do arquivo:")
print("-" * 40)

with open('arquivos/relatorio.txt', 'r', encoding='utf-8') as arquivo:
    for linha in arquivo:
        linha_limpa = linha.strip()  # .strip() remove \n e espaços extras
        if linha_limpa:  # ignora linhas vazias
            print(f"  → {linha_limpa}")
````

### ➕ 1.3 — Adicionando conteúdo (modo append)

> **Atenção!** O modo `"w"` **apaga** o conteúdo anterior.  
> Use o modo `"a"` (append) para **adicionar** ao final sem perder o que já existe.

````python
# Simulando um log que recebe novos eventos ao longo do dia
novas_linhas = [
    "\n--- ATUALIZAÇÃO 16h30 ---",
    "Nova venda registrada: Notebook Dell - R$ 3.500,00",
    "Estoque atualizado: 9 unidades restantes",
]

# Modo "a" = append (acrescenta ao final)
with open('arquivos/relatorio.txt', 'a', encoding='utf-8') as arquivo:
    for linha in novas_linhas:
        arquivo.write(linha + '\n')

print('✅ Conteúdo adicionado ao arquivo!')

# Verificando o resultado
print("\n📄 Arquivo completo após append:")
print("-" * 40)
with open('arquivos/relatorio.txt', 'r', encoding='utf-8') as arquivo:
    print(arquivo.read())
````

### 🔍 1.4 — Métodos avançados: tell() e seek()

Esses métodos permitem navegar dentro do arquivo como se fosse uma fita.

````python
with open('arquivos/relatorio.txt', 'r', encoding='utf-8') as arquivo:
    
    # Lendo os primeiros 30 caracteres
    trecho = arquivo.read(30)
    print(f"Primeiros 30 caracteres: '{trecho}'")
    
    # tell() mostra a posição atual do cursor no arquivo
    posicao = arquivo.tell()
    print(f"Posição atual do cursor: {posicao}")
    
    # seek(0, 0) retorna o cursor para o início do arquivo
    arquivo.seek(0, 0)
    print(f"Posição após seek(0): {arquivo.tell()}")
    
    # Agora podemos ler novamente do início
    primeira_linha = arquivo.readline()
    print(f"Primeira linha: {primeira_linha.strip()}")
````

# 📊 PARTE 2 — Arquivos CSV

CSV = **Comma-Separated Values** (Valores Separados por Vírgula)

É o formato mais usado para troca de dados tabulares. Pense nele como uma planilha do Excel em formato texto.

**🧠 Analogia:** O CSV é o "idioma universal" dos dados. O Excel usa, o Google Sheets usa, o Python usa, os bancos de dados usam. É a forma mais simples de mover dados entre sistemas diferentes.

**Exemplo de arquivo CSV:**
```
nome,cidade,vendas,receita
Ana Costa,Fortaleza,15,45000.00
Bruno Lima,São Paulo,8,24000.00
Carla Dias,Recife,22,66000.00
```


### ✏️ 2.1 — Criando um arquivo CSV

````python
import csv

# Método 1: csv.writer — escrevendo linha por linha como lista
# newline='' é IMPORTANTE no Windows para evitar linhas em branco extras

with open('arquivos/numeros.csv', 'w', newline='', encoding='utf-8') as arquivo:
    escritor = csv.writer(arquivo)
    
    # Cabeçalho
    escritor.writerow(['primeira', 'segunda', 'terceira'])
    
    # Linhas de dados
    escritor.writerow([55, 93, 76])
    escritor.writerow([62, 14, 86])

print('✅ Arquivo "numeros.csv" criado!')

# Visualizando o conteúdo do arquivo
print("\nConteúdo do arquivo:")
with open('arquivos/numeros.csv', 'r') as f:
    print(f.read())
````

````python
# Método 2: csv.DictWriter — escrevendo com nomes de colunas
# RECOMENDADO: mais legível e menos propenso a erros

vendedores = [
    {"nome": "Ana Costa",  "cidade": "Fortaleza", "vendas": 15, "receita": 45000.00},
    {"nome": "Bruno Lima", "cidade": "São Paulo",  "vendas":  8, "receita": 24000.00},
    {"nome": "Carla Dias", "cidade": "Recife",     "vendas": 22, "receita": 66000.00},
]

campos = ["nome", "cidade", "vendas", "receita"]

with open('arquivos/vendedores.csv', 'w', newline='', encoding='utf-8') as arquivo:
    escritor = csv.DictWriter(arquivo, fieldnames=campos)
    escritor.writeheader()        # escreve a linha de cabeçalho
    escritor.writerows(vendedores) # escreve todas as linhas de uma vez

print('✅ Arquivo "vendedores.csv" criado!')
with open('arquivos/vendedores.csv', 'r') as f:
    print("\nConteúdo:")
    print(f.read())
````

### 📖 2.2 — Lendo um arquivo CSV

````python
# Método 1: csv.reader — cada linha vira uma LISTA
# Você acessa por posição: linha[0], linha[1], etc.

print("Usando csv.reader (acesso por índice):")
print("-" * 40)

with open('arquivos/vendedores.csv', 'r', encoding='utf-8') as arquivo:
    leitor = csv.reader(arquivo)
    for i, linha in enumerate(leitor):
        if i == 0:
            print(f"CABEÇALHO: {linha}")
        else:
            print(f"  Nome: {linha[0]} | Cidade: {linha[1]} | Vendas: {linha[2]}")
````

````python
# Método 2: csv.DictReader — cada linha vira um DICIONÁRIO
# RECOMENDADO: você acessa pelo nome da coluna

print("Usando csv.DictReader (acesso por nome de coluna):")
print("-" * 50)

total_receita = 0

with open('arquivos/vendedores.csv', 'r', encoding='utf-8') as arquivo:
    leitor = csv.DictReader(arquivo)
    for linha in leitor:
        receita = float(linha["receita"])
        total_receita += receita
        print(f"  {linha['nome']} ({linha['cidade']}): {linha['vendas']} vendas | R$ {receita:,.2f}")

print("-" * 50)
print(f"  TOTAL DE RECEITA: R$ {total_receita:,.2f}")
````

````python
# Convertendo para lista — útil para manipular os dados depois

with open('arquivos/vendedores.csv', 'r', encoding='utf-8', newline='') as arquivo:
    leitor = csv.reader(arquivo)
    dados = list(leitor)

print(f"Todos os dados como lista: {len(dados)} itens")
print(f"Cabeçalho: {dados[0]}")
print(f"Primeira linha de dados: {dados[1]}")
print(f"\nAcessando célula específica:")
print(f"  dados[1][0] = '{dados[1][0]}'  (nome do 1º vendedor)")
print(f"  dados[2][3] = '{dados[2][3]}'  (receita do 2º vendedor)")
````

# 🔗 PARTE 3 — Arquivos JSON

JSON = **JavaScript Object Notation**

É o formato padrão para troca de dados na web. Toda API REST retorna JSON. É como um dicionário Python, mas em texto.

**🧠 Analogia:** O CSV é uma planilha simples. O JSON é um formulário com sub-formulários. Ele permite representar dados complexos (um cliente com vários endereços, um pedido com vários itens) de forma natural.

### Tipos de dados em JSON

| JSON | Python |
|---|---|
| `{"chave": valor}` | `dict` |
| `[item1, item2]` | `list` |
| `"texto"` | `str` |
| `42` ou `3.14` | `int` ou `float` |
| `true` / `false` | `True` / `False` |
| `null` | `None` |


### ✏️ 3.1 — Criando e salvando um JSON

````python
import json

# Criando um dicionário Python com dados de um desenvolvedor famoso
# (Guido van Rossum é o criador do Python!)
desenvolvedor = {
    'nome': 'Guido van Rossum',
    'linguagem': 'Python',
    'linguagens_inspiracao': ['C', 'Modula-3', 'Lisp'],
    'usuarios_estimados': 10_000_000,
    'ativo': True
}

print("Dicionário Python:")
for chave, valor in desenvolvedor.items():
    print(f"  {chave}: {valor}")
````

````python
# Convertendo dicionário → string JSON (json.dumps = dump to String)
json_string = json.dumps(desenvolvedor, ensure_ascii=False, indent=4)

print("Representação em JSON:")
print(json_string)
````

````python
# Salvando em arquivo (json.dump = dump to File)
# ensure_ascii=False → mantém acentos (ã, ç, é) sem converter para \u00e3
# indent=4 → formata com indentação, deixando o arquivo legível

with open('arquivos/desenvolvedor.json', 'w', encoding='utf-8') as arquivo:
    json.dump(desenvolvedor, arquivo, ensure_ascii=False, indent=4)

print('✅ Arquivo "desenvolvedor.json" criado!')
````

### 📖 3.2 — Lendo e navegando em um JSON

````python
# Lendo de arquivo (json.load = load from File)

with open('arquivos/desenvolvedor.json', 'r', encoding='utf-8') as arquivo:
    dados = json.load(arquivo)

print(f"Tipo dos dados carregados: {type(dados)}")
print(f"\nAcessando valores:")
print(f"  Nome: {dados['nome']}")
print(f"  Linguagem: {dados['linguagem']}")
print(f"  Está ativo? {dados['ativo']}")
print(f"  Usuários estimados: {dados['usuarios_estimados']:,}")

print(f"\nPercorrendo lista de inspirações:")
for linguagem in dados['linguagens_inspiracao']:
    print(f"  → {linguagem}")
````

### 🏢 3.3 — Exemplo de negócio: JSON com estrutura complexa

````python
# Simulando um pedido de e-commerce (como viria de uma API)
pedido = {
    "pedido_id": 1042,
    "cliente": {
        "nome": "Ana Costa",
        "email": "ana@email.com",
        "cidade": "Fortaleza"
    },
    "itens": [
        {"produto": "Notebook Dell",   "quantidade": 1, "preco": 3500.00},
        {"produto": "Mouse Logitech", "quantidade": 2, "preco":  150.00}
    ],
    "status": "aprovado",
    "pago": True
}

# Salvando
with open('arquivos/pedido.json', 'w', encoding='utf-8') as f:
    json.dump(pedido, f, ensure_ascii=False, indent=4)

print('✅ Pedido salvo!')
````

````python
# Lendo e processando o pedido
with open('arquivos/pedido.json', 'r', encoding='utf-8') as f:
    pedido = json.load(f)

print(f"=== PEDIDO #{pedido['pedido_id']} ===")
print(f"Cliente : {pedido['cliente']['nome']} ({pedido['cliente']['cidade']})")
print(f"Status  : {pedido['status'].upper()}")
print(f"Pago    : {'✅ Sim' if pedido['pago'] else '❌ Não'}")
print()
print("Itens do pedido:")

total = 0
for item in pedido['itens']:
    subtotal = item['quantidade'] * item['preco']
    total += subtotal
    print(f"  {item['quantidade']}x {item['produto']:20} = R$ {subtotal:,.2f}")

print("-" * 45)
print(f"  TOTAL DO PEDIDO             = R$ {total:,.2f}")
````

### 🌐 3.4 — Copiando conteúdo entre arquivos

````python
# Copiando o conteúdo de um arquivo JSON para um arquivo TXT
# Útil quando você precisa converter ou mover dados

arquivo_origem  = 'arquivos/desenvolvedor.json'
arquivo_destino = 'arquivos/desenvolvedor_backup.txt'

# Método recomendado: with aninhado
with open(arquivo_origem, 'r', encoding='utf-8') as entrada:
    conteudo = entrada.read()
    with open(arquivo_destino, 'w', encoding='utf-8') as saida:
        saida.write(conteudo)

print(f'✅ Arquivo copiado: {arquivo_origem} → {arquivo_destino}')

# Verificando
with open(arquivo_destino, 'r') as f:
    print(f"\nConteúdo do backup ({len(f.read())} caracteres):")

with open(arquivo_destino, 'r') as f:
    print(f.read())
````

## 📚 Resumo da Aula

| Operação | TXT | CSV | JSON |
|---|---|---|---|
| **Abrir para leitura** | `open(f, 'r')` | `open(f, 'r')` | `open(f, 'r')` |
| **Ler conteúdo** | `.read()` / `.readlines()` | `csv.reader()` / `DictReader()` | `json.load()` |
| **Criar/Sobrescrever** | `open(f, 'w')` + `.write()` | `csv.writer()` / `DictWriter()` | `json.dump()` |
| **Acrescentar** | `open(f, 'a')` + `.write()` | *(reescrever é mais seguro)* | *(reescrever é mais seguro)* |
| **Estrutura dos dados** | Texto livre | Tabular (linhas e colunas) | Hierárquica (aninhada) |
| **Caso de uso típico** | Logs, relatórios | Planilhas, exportações | APIs, configurações |

### ✅ Boas práticas
1. **Sempre use `with open(...)`** — fecha o arquivo automaticamente
2. **Sempre use `encoding='utf-8'`** — evita problemas com acentos
3. **CSV no Windows: use `newline=''`** — evita linhas em branco extras
4. **JSON: use `ensure_ascii=False, indent=4`** — mantém acentos e fica legível
5. **Prefira `DictReader`/`DictWriter`** — código mais claro e menos erros
