# 🤔 PARTE 1 — Por que Programação Orientada a Objetos?

## O problema que a POO resolve

Imagine que você precisa gerenciar os dados de vários funcionários na sua empresa.  
Sem POO, você faria assim:


````python
# ❌ JEITO ANTIGO — Programação Estruturada (sem POO)
# Para 2 funcionários já fica confuso. Imagine com 100!

func1_nome   = "Ana Costa"
func1_cargo  = "Analista de Dados"
func1_salario = 7500.00

func2_nome   = "Bruno Lima"
func2_cargo  = "Engenheiro de Dados"
func2_salario = 9000.00

# Para dar aumento precisamos saber o nome da variável exata
func1_salario = func1_salario * 1.10  # 10% de aumento
func2_salario = func2_salario * 1.10

print(f"{func1_nome}: R$ {func1_salario:,.2f}")
print(f"{func2_nome}: R$ {func2_salario:,.2f}")

print("\n⚠️ Problemas desta abordagem:")
print("  - Variáveis soltas e sem organização")
print("  - Com 100 funcionários seriam 300 variáveis!")
print("  - Difícil de manter e expandir")
````

````python
# ✅ JEITO POO — Organizado, escalável e legível
# Vamos construir essa classe ao longo do notebook

class Funcionario:
    def __init__(self, nome, cargo, salario):
        self.nome    = nome
        self.cargo   = cargo
        self.salario = salario

    def dar_aumento(self, percentual):
        self.salario *= (1 + percentual / 100)
        print(f"✅ {self.nome} recebeu {percentual}% de aumento! Novo salário: R$ {self.salario:,.2f}")

# Com POO: criamos quantos funcionários quisermos da mesma forma
func1 = Funcionario("Ana Costa",  "Analista de Dados",    7500.00)
func2 = Funcionario("Bruno Lima", "Engenheiro de Dados",  9000.00)
func3 = Funcionario("Carla Dias", "Cientista de Dados",  11000.00)

func1.dar_aumento(10)
func2.dar_aumento(10)
func3.dar_aumento(10)

print("\n✅ Vantagens da POO:")
print("  - Dados e comportamentos organizados juntos")
print("  - Funciona igual para qualquer número de funcionários")
print("  - Fácil de manter, expandir e reutilizar")
````

# 🏗️ PARTE 2 — Classe e Objeto: o Molde e as Instâncias

## A analogia da ficha cadastral

```
┌─────────────────────────────────────┐
│           CLASSE: Cliente           │  ← O MOLDE (a ficha em branco)
│─────────────────────────────────────│
│  nome   : ___________________       │
│  email  : ___________________       │
│  cidade : ___________________       │
│─────────────────────────────────────│
│  fazer_compra()                     │
│  exibir_info()                      │
└─────────────────────────────────────┘
          ↓ instanciar ↓
┌──────────────────┐  ┌──────────────────┐
│  OBJETO: cliente1│  │  OBJETO: cliente2│  ← As fichas preenchidas
│──────────────────│  │──────────────────│
│  nome: Ana Costa │  │  nome: Bruno Lima│
│  email: ana@...  │  │  email: bruno@...│
│  cidade: Fortlz  │  │  cidade: SP      │
└──────────────────┘  └──────────────────┘
```

- **Classe** = o modelo, o projeto, o molde. Define *como* algo deve ser.
- **Objeto** = uma instância concreta criada a partir da classe. É *o que existe de verdade*.

````python
# A classe mais simples possível — só o molde vazio
# "pass" significa "não faz nada por enquanto"

class Cliente:
    pass

# Criando objetos (instâncias) da classe
cliente1 = Cliente()  # instancia o objeto
cliente2 = Cliente()  # outro objeto diferente

print(f"Tipo de cliente1: {type(cliente1)}")
print(f"Tipo de cliente2: {type(cliente2)}")
print(f"São o mesmo objeto? {cliente1 is cliente2}")  # False! São objetos diferentes
print(f"ID na memória de cliente1: {id(cliente1)}")
print(f"ID na memória de cliente2: {id(cliente2)}")
````

# 🏷️ PARTE 3 — Atributos: as Características do Objeto

Atributos são as **informações** que cada objeto guarda.  
São como os campos de uma ficha cadastral.

## O método `__init__` (construtor)

É chamado automaticamente quando criamos um objeto.  
Serve para inicializar os atributos com valores.

```python
def __init__(self, parametro1, parametro2):
#             ↑
#          sempre o 1º parâmetro
#          representa o PRÓPRIO objeto
```

## O parâmetro `self`

`self` é uma referência ao **próprio objeto** que está sendo criado ou manipulado.  
É como se o objeto dissesse: *"Eu mesmo, o objeto que está aqui."*

````python
class Cliente:
    """
    Representa um cliente da nossa loja.
    
    Atributos:
        nome    (str) : nome completo do cliente
        email   (str) : e-mail para contato
        cidade  (str) : cidade onde mora
        ativo   (bool): se o cadastro está ativo
    """
    
    def __init__(self, nome, email, cidade, ativo=True):
        #            ↑ self é sempre o 1º parâmetro
        self.nome   = nome    # guarda 'nome' como atributo do objeto
        self.email  = email   # cada objeto tem seu próprio .email
        self.cidade = cidade  # cada objeto tem seu próprio .cidade
        self.ativo  = ativo   # valor padrão = True (opcional)

# Criando objetos — o Python chama __init__ automaticamente
cliente1 = Cliente("Ana Costa",  "ana@email.com",   "Fortaleza")
cliente2 = Cliente("Bruno Lima", "bruno@email.com", "São Paulo")
cliente3 = Cliente("Carla Dias", "carla@email.com", "Recife", ativo=False)

# Acessando atributos com ponto (objeto.atributo)
print(f"Nome do cliente1 : {cliente1.nome}")
print(f"E-mail do cliente2: {cliente2.email}")
print(f"Cliente3 ativo?   : {cliente3.ativo}")

print("\nCada objeto tem seus PRÓPRIOS atributos:")
print(f"  cliente1.cidade = '{cliente1.cidade}'")
print(f"  cliente2.cidade = '{cliente2.cidade}'")
print(f"  cliente3.cidade = '{cliente3.cidade}'")
````

````python
# Atributos podem ser modificados depois da criação

print(f"Antes: {cliente3.nome} está ativo? {cliente3.ativo}")

cliente3.ativo = True  # reativa o cadastro
cliente3.email = "carla.nova@email.com"  # atualiza o e-mail

print(f"Depois: {cliente3.nome} está ativo? {cliente3.ativo}")
print(f"Novo e-mail: {cliente3.email}")
````

# ⚙️ PARTE 4 — Métodos: os Comportamentos do Objeto

Métodos são **funções definidas dentro de uma classe**.  
Representam as **ações** que um objeto pode realizar.

```
Classe Produto:
  Atributos → O QUE O PRODUTO É    (nome, preço, estoque)
  Métodos   → O QUE O PRODUTO FAZ  (vender, repor, exibir_info)
```

> **Regra:** Todo método de instância sempre tem `self` como primeiro parâmetro.

````python
class Produto:
    """
    Representa um produto do estoque de uma loja.
    
    Atributos: nome, categoria, preco, estoque
    Métodos  : vender(), repor_estoque(), aplicar_desconto(), info()
    """

    def __init__(self, nome, categoria, preco, estoque=0):
        self.nome      = nome
        self.categoria = categoria
        self.preco     = preco
        self.estoque   = estoque

    # ── Método 1: exibir informações ──────────────────────────────
    def info(self):
        """Exibe as informações do produto formatadas."""
        status = f"{self.estoque} un." if self.estoque > 0 else "❌ SEM ESTOQUE"
        print(f"📦 {self.nome:25} | {self.categoria:15} | R$ {self.preco:8.2f} | {status}")

    # ── Método 2: realizar uma venda ──────────────────────────────
    def vender(self, quantidade):
        """Registra uma venda e atualiza o estoque."""
        if quantidade <= 0:
            print("❌ Quantidade inválida.")
        elif quantidade > self.estoque:
            print(f"❌ Estoque insuficiente! Temos apenas {self.estoque} unidade(s) de '{self.nome}'.")
        else:
            self.estoque -= quantidade
            total = quantidade * self.preco
            print(f"✅ Venda realizada: {quantidade}x '{self.nome}' = R$ {total:,.2f} | Estoque restante: {self.estoque}")
            return total  # retorna o valor para uso externo

    # ── Método 3: repor estoque ───────────────────────────────────
    def repor_estoque(self, quantidade):
        """Adiciona unidades ao estoque."""
        if quantidade > 0:
            self.estoque += quantidade
            print(f"📥 Estoque reposto: +{quantidade} un. de '{self.nome}'. Total: {self.estoque} un.")
        else:
            print("❌ Quantidade de reposição deve ser positiva.")

    # ── Método 4: aplicar desconto ────────────────────────────────
    def aplicar_desconto(self, percentual):
        """Aplica um desconto percentual no preço."""
        if 0 < percentual < 100:
            preco_antigo = self.preco
            self.preco  *= (1 - percentual / 100)
            print(f"🏷️  Desconto de {percentual}% em '{self.nome}': R$ {preco_antigo:.2f} → R$ {self.preco:.2f}")
        else:
            print("❌ Percentual deve estar entre 0 e 100.")


# ─── Testando a classe Produto ────────────────────────────────────
print("=" * 65)
print("CATÁLOGO DE PRODUTOS")
print("=" * 65)

notebook = Produto("Notebook Dell Inspiron", "Informática", 3500.00, 10)
mouse    = Produto("Mouse Logitech MX3",     "Informática",  280.00, 50)
cadeira  = Produto("Cadeira Gamer Pro",       "Móveis",      1200.00,  5)

notebook.info()
mouse.info()
cadeira.info()
````

````python
# Simulando operações do dia a dia

print("\n🛒 SIMULAÇÃO DE VENDAS\n")

notebook.vender(3)       # ✅ Vende 3 notebooks
mouse.vender(100)        # ❌ Não tem 100 no estoque
mouse.vender(15)         # ✅ Vende 15 mouses
cadeira.vender(5)        # ✅ Vende todas as cadeiras
cadeira.vender(1)        # ❌ Sem estoque

print("\n📥 REPOSIÇÃO DE ESTOQUE\n")
cadeira.repor_estoque(20)

print("\n🏷️  PROMOÇÃO DE VENDAS\n")
notebook.aplicar_desconto(15)  # 15% off
mouse.aplicar_desconto(5)      # 5% off

print("\n📊 ESTOQUE ATUAL\n")
print(f"{'Produto':25} | {'Categoria':15} | {'Preço':>10} | Estoque")
print("-" * 65)
notebook.info()
mouse.info()
cadeira.info()
````

# 🏦 PARTE 5 — Exemplo Completo: Conta Bancária

Vamos construir uma classe mais completa que simula uma conta bancária.  
Este é um exemplo clássico que demonstra como a POO organiza dados e comportamentos juntos.

**🧠 Analogia de negócio:**  
Uma conta bancária tem informações (titular, saldo, número) e ações (depositar, sacar, transferir, ver extrato). Com POO, tudo isso fica organizado em um único lugar — a classe `ContaBancaria`.

````python
from datetime import datetime

class ContaBancaria:
    """
    Simula uma conta bancária com histórico de transações.
    
    Atributos:
        titular     : nome do dono da conta
        numero      : número da conta
        saldo       : saldo atual (padrão = 0)
        historico   : lista com todas as transações
    """

    def __init__(self, titular, numero, saldo_inicial=0):
        self.titular   = titular
        self.numero    = numero
        self.saldo     = saldo_inicial
        self.historico = []  # lista vazia — vai crescendo

        # Registra a abertura no histórico
        if saldo_inicial > 0:
            self._registrar("Saldo inicial", saldo_inicial)

    # ── Método privado (convenção: _ antes do nome) ────────────────
    def _registrar(self, tipo, valor):
        """Registra uma transação no histórico (uso interno)."""
        agora = datetime.now().strftime("%d/%m/%Y %H:%M")
        self.historico.append({
            "data": agora,
            "tipo": tipo,
            "valor": valor,
            "saldo_apos": self.saldo
        })

    # ── Depositar ─────────────────────────────────────────────────
    def depositar(self, valor):
        """Adiciona valor ao saldo."""
        if valor <= 0:
            print("❌ O valor do depósito deve ser positivo.")
            return
        self.saldo += valor
        self._registrar("Depósito", valor)
        print(f"✅ Depósito de R$ {valor:,.2f} realizado. Saldo: R$ {self.saldo:,.2f}")

    # ── Sacar ─────────────────────────────────────────────────────
    def sacar(self, valor):
        """Retira valor do saldo se houver saldo suficiente."""
        if valor <= 0:
            print("❌ O valor do saque deve ser positivo.")
            return False
        if valor > self.saldo:
            print(f"❌ Saldo insuficiente! Saldo atual: R$ {self.saldo:,.2f}")
            return False
        self.saldo -= valor
        self._registrar("Saque", -valor)
        print(f"💸 Saque de R$ {valor:,.2f} realizado. Saldo: R$ {self.saldo:,.2f}")
        return True

    # ── Transferir ────────────────────────────────────────────────
    def transferir(self, valor, conta_destino):
        """Transfere valor para outra conta."""
        print(f"🔄 Transferindo R$ {valor:,.2f} para {conta_destino.titular}...")
        if self.sacar(valor):
            conta_destino.depositar(valor)
            self._registrar(f"Transferência para {conta_destino.titular}", -valor)

    # ── Extrato ───────────────────────────────────────────────────
    def extrato(self):
        """Exibe o extrato completo da conta."""
        print("\n" + "═" * 55)
        print(f"  EXTRATO — {self.titular}")
        print(f"  Conta: {self.numero}")
        print("═" * 55)
        if not self.historico:
            print("  Sem transações registradas.")
        else:
            for t in self.historico:
                sinal = "+" if t["valor"] >= 0 else ""
                print(f"  {t['data']}  |  {t['tipo']:30} | {sinal}R$ {t['valor']:>10,.2f}")
        print("─" * 55)
        print(f"  SALDO ATUAL: R$ {self.saldo:,.2f}")
        print("═" * 55 + "\n")


# ─── Testando a Conta Bancária ────────────────────────────────────
conta_ana   = ContaBancaria("Ana Costa",  "001-2345", saldo_inicial=5000.00)
conta_bruno = ContaBancaria("Bruno Lima", "001-6789", saldo_inicial=1200.00)

print("✅ Contas criadas com sucesso!")
print(f"   Ana   → saldo inicial: R$ {conta_ana.saldo:,.2f}")
print(f"   Bruno → saldo inicial: R$ {conta_bruno.saldo:,.2f}")
````

````python
# Simulando um dia de movimentações bancárias

print("\n📅 MOVIMENTAÇÕES DO DIA\n")

conta_ana.depositar(1500.00)          # recebe salário
conta_ana.sacar(300.00)               # paga conta
conta_ana.transferir(500.00, conta_bruno)  # envia para Bruno
conta_ana.sacar(10000.00)             # tenta sacar mais do que tem

conta_bruno.depositar(800.00)         # Bruno recebe outro depósito
conta_bruno.sacar(200.00)             # Bruno saca
````

````python
# Visualizando os extratos
conta_ana.extrato()
conta_bruno.extrato()
````

# 🛒 PARTE 6 — Exemplo Completo: Sistema de Pedidos

Vamos agora combinar **duas classes que se relacionam**: `Produto` e `Pedido`.  
Isso é muito próximo de como sistemas reais de e-commerce funcionam.

```
Produto ──────────────────────────────────────────────
  nome, preco, estoque                                 
  vender(), repor(), info()                            
                                                       
Pedido ───────────────────────────────────────────────
  cliente, itens[], status                             
  adicionar_item(), calcular_total(), finalizar()      
```

````python
class Pedido:
    """
    Representa um pedido de compra.
    Relaciona-se com objetos da classe Produto.
    """

    _contador = 0  # atributo de CLASSE — compartilhado entre todos os pedidos

    def __init__(self, nome_cliente):
        Pedido._contador += 1
        self.numero_pedido = Pedido._contador
        self.nome_cliente  = nome_cliente
        self.itens         = []  # lista de dicionários {produto, quantidade, subtotal}
        self.status        = "aberto"

    def adicionar_item(self, produto, quantidade):
        """Adiciona um produto ao pedido verificando o estoque."""
        if quantidade <= 0:
            print("❌ Quantidade deve ser positiva.")
            return
        if quantidade > produto.estoque:
            print(f"❌ '{produto.nome}': pediu {quantidade} mas só tem {produto.estoque} em estoque.")
            return
        subtotal = quantidade * produto.preco
        self.itens.append({
            "produto":    produto,
            "quantidade": quantidade,
            "subtotal":   subtotal
        })
        print(f"  ➕ {quantidade}x {produto.nome} adicionado (R$ {subtotal:,.2f})")

    def calcular_total(self):
        """Retorna o valor total do pedido."""
        return sum(item["subtotal"] for item in self.itens)

    def finalizar(self):
        """Confirma o pedido e baixa o estoque de cada produto."""
        if not self.itens:
            print("❌ Pedido vazio — adicione itens primeiro.")
            return
        if self.status != "aberto":
            print(f"❌ Este pedido já está '{self.status}'.")
            return
        # Baixa o estoque de cada produto
        for item in self.itens:
            item["produto"].estoque -= item["quantidade"]
        self.status = "finalizado"
        print(f"✅ Pedido #{self.numero_pedido} finalizado! Total: R$ {self.calcular_total():,.2f}")

    def exibir(self):
        """Exibe o resumo do pedido."""
        print(f"\n{'═'*55}")
        print(f"  PEDIDO #{self.numero_pedido:04d} — {self.status.upper()}")
        print(f"  Cliente: {self.nome_cliente}")
        print(f"{'─'*55}")
        if not self.itens:
            print("  (nenhum item)")
        else:
            for item in self.itens:
                print(f"  {item['quantidade']:>3}x  {item['produto'].nome:25} R$ {item['subtotal']:>9,.2f}")
        print(f"{'─'*55}")
        print(f"  TOTAL: R$ {self.calcular_total():>41,.2f}")
        print(f"{'═'*55}\n")


print("✅ Classe Pedido definida!")
````

````python
# Recriando nossos produtos
notebook = Produto("Notebook Dell",     "Informática", 3500.00, 10)
mouse    = Produto("Mouse Logitech",    "Informática",  280.00, 50)
cadeira  = Produto("Cadeira Gamer Pro", "Móveis",      1200.00,  5)
fone     = Produto("Fone Sony WH-1000", "Informática",  900.00, 20)

# Criando pedidos
print("🛒 Montando pedido da Ana...")
pedido_ana = Pedido("Ana Costa")
pedido_ana.adicionar_item(notebook, 1)
pedido_ana.adicionar_item(mouse,    2)
pedido_ana.adicionar_item(fone,     1)

print("\n🛒 Montando pedido do Bruno...")
pedido_bruno = Pedido("Bruno Lima")
pedido_bruno.adicionar_item(cadeira, 2)
pedido_bruno.adicionar_item(mouse,   1)
pedido_bruno.adicionar_item(cadeira, 10)  # ❌ sem estoque suficiente
````

````python
# Finalizando os pedidos
print("📋 Resumo dos pedidos antes de finalizar:")
pedido_ana.exibir()
pedido_bruno.exibir()

print("🏁 Finalizando pedidos...")
pedido_ana.finalizar()
pedido_bruno.finalizar()

print("\n📦 Estoque após os pedidos:")
print(f"  Notebook : {notebook.estoque} un.")
print(f"  Mouse    : {mouse.estoque} un.")
print(f"  Cadeira  : {cadeira.estoque} un.")
print(f"  Fone     : {fone.estoque} un.")

print(f"\n  Total de pedidos criados: {Pedido._contador}")
````

# 📌 PARTE 7 — Conceitos Importantes

## 7.1 — Métodos que retornam valores vs. que só imprimem

| Tipo | Usa `return` | Exemplo |
|---|---|---|
| Só exibe | Não | `def info(self): print(...)` |
| Retorna valor | Sim | `def calcular_total(self): return sum(...)` |

Métodos com `return` podem ser usados em expressões e guardados em variáveis.

````python
# Demonstrando return vs. print

class Retangulo:
    """Representa um retângulo com largura e altura."""

    def __init__(self, largura, altura):
        self.largura = largura
        self.altura  = altura

    def area(self):
        """RETORNA a área — podemos usar o valor depois."""
        return self.largura * self.altura

    def perimetro(self):
        """RETORNA o perímetro."""
        return 2 * (self.largura + self.altura)

    def e_quadrado(self):
        """RETORNA True se for quadrado."""
        return self.largura == self.altura

    def descrever(self):
        """Só IMPRIME — não retorna nada útil."""
        print(f"Retângulo {self.largura}x{self.altura}")
        print(f"  Área      : {self.area()} m²")        # usa o método area()
        print(f"  Perímetro : {self.perimetro()} m")
        print(f"  É quadrado: {self.e_quadrado()}")


r1 = Retangulo(4, 6)
r2 = Retangulo(5, 5)

# Usando métodos com return em expressões
area_total = r1.area() + r2.area()
print(f"Área total dos dois retângulos: {area_total} m²")

if r2.e_quadrado():
    print(f"r2 é um quadrado de lado {r2.largura}m!")

print()
r1.descrever()
print()
r2.descrever()
````

## 7.2 — Atributo de Classe vs. Atributo de Instância

| Tipo | Definido | Compartilhado? | Exemplo |
|---|---|---|---|
| **Instância** | No `__init__` com `self.` | ❌ Cada objeto tem o seu | `self.nome`, `self.saldo` |
| **Classe** | Fora do `__init__`, direto na classe | ✅ Todos os objetos compartilham | `Pedido._contador` |

````python
class Empresa:
    # Atributo de CLASSE — compartilhado por todos os objetos
    pais          = "Brasil"
    total_empresas = 0

    def __init__(self, nome, setor):
        # Atributos de INSTÂNCIA — cada empresa tem os seus
        self.nome  = nome
        self.setor = setor
        Empresa.total_empresas += 1  # atualiza o contador da CLASSE

    def info(self):
        # self.pais acessa o atributo de classe via instância (funciona!)
        print(f"🏢 {self.nome} | Setor: {self.setor} | País: {self.pais}")


emp1 = Empresa("TechData",  "Tecnologia")
emp2 = Empresa("AgroSmart", "Agronegócio")
emp3 = Empresa("FinBank",   "Financeiro")

emp1.info()
emp2.info()
emp3.info()

print(f"\nTotal de empresas cadastradas: {Empresa.total_empresas}")
print(f"País (via classe) : {Empresa.pais}")
print(f"País (via objeto) : {emp1.pais}")
````

### 🗝️ Conceitos-chave

| Conceito | O que é | Exemplo |
|---|---|---|
| **Classe** | O molde, o projeto | `class Produto:` |
| **Objeto** | Instância concreta | `p = Produto(...)` |
| **`__init__`** | Construtor (chamado na criação) | `def __init__(self, nome):` |
| **`self`** | Referência ao próprio objeto | `self.nome = nome` |
| **Atributo de instância** | Dado específico de cada objeto | `self.saldo` |
| **Atributo de classe** | Dado compartilhado por todos | `Pedido._contador` |
| **Método** | Função da classe | `def depositar(self, valor):` |
| **`return`** | Devolve um valor para uso externo | `return self.saldo` |

### ✅ Boas práticas

1. **Nome de classe sempre em CamelCase** — `ContaBancaria`, não `conta_bancaria`
2. **Docstring em toda classe e método** — explica o que faz
3. **`__init__` inicializa todos os atributos** — sem atributos surpresa depois
4. **Métodos fazem UMA coisa bem feita** — `depositar()` só deposita
5. **`_nome_privado`** — convenção para métodos/atributos de uso interno
6. **Prefira `return` a `print`** em métodos que calculam — deixa o código reutilizável

### 🧠 A grande analogia do negócio

```
Classe      ↔   Ficha cadastral em branco
Objeto      ↔   Ficha preenchida com dados reais
__init__    ↔   O momento de preencher a ficha
Atributos   ↔   Os campos da ficha (nome, CPF, cidade)
Métodos     ↔   As ações que o cadastrado pode fazer (comprar, solicitar, cancelar)
self        ↔   "Este cliente específico aqui"
```
