Dada a seguinte string:
texto = "O rato roeu a roupa do rei de Roma"
1 - Crie uma nova string com todas as letras maiúsculas.
2 - Crie uma nova string com todas as letras minúsculas.
3 - Conte quantas vezes a letra "a" aparece na string.
4 - Substitua a palavra "rei" por "rainha" na string.
5 - Divida a string em uma lista de palavras.
6 - Ordene a lista de palavras em ordem alfabética.

```python
# Defina a string original
texto = "O rato roeu a roupa do rei de Roma"

# 1. Crie uma nova string com todas as letras maiúsculas.
texto_maiusculo = texto.upper()
print("Texto em maiúsculas: ", texto_maiusculo)

# 2. Crie uma nova string com todas as letras minúsculas.
texto_minusculo = texto.lower()
print("Texto em minúsculas: ", texto_minusculo)

# 3. Conte quantas vezes a letra "a" aparece na string.
quantidade_a = texto.count('a')
print("Quantidade de letras 'a' no texto: ", quantidade_a)

# 4. Substitua a palavra "rei" por "rainha" na string.
texto_rainha = texto.replace('rei', 'rainha')
print("Texto com a palavra 'rei' substituída por 'rainha': ", texto_rainha)

# 5. Divida a string em uma lista de palavras.
palavras = texto.split()
print("Lista de palavras: ", palavras)

# 6. Ordene a lista de palavras em ordem alfabética.
palavras.sort()
print("Lista de palavras ordenada: ", palavras)

# 7. Junte a lista de palavras novamente em uma única string, separando cada palavra por um traço ("-").
texto_separado_por_traco = '-'.join(palavras)
print("Texto separado por traço: ", texto_separado_por_traco)
```

```python
# 6. Ordene a lista de palavras em ordem alfabética.
palavras.sort()
print("Lista de palavras ordenada: ", palavras)
```
