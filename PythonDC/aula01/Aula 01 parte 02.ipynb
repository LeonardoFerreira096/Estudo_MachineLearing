```python
import datetime

# Criando uma data
d = datetime.date(2022, 2, 10)
print(d)  # 2022-02-10

# Obtendo o dia da semana (0 = segunda-feira, 6 = domingo)
print(d.weekday())  # 3 (quinta-feira)

# Obtendo o ano, mês e dia separadamente
print(d.year)  # 2022
print(d.month)  # 2
print(d.day)  # 10

# Criando um horário
t = datetime.time(14, 30, 0)
print(t)  # 14:30:00

# Obtendo a hora, minuto e segundo separadamente
print(t.hour)  # 14
print(t.minute)  # 30
print(t.second)  # 0

# Criando uma data/hora completa
dt = datetime.datetime(2022, 2, 10, 14, 30, 0)
print(dt)  # 2022-02-10 14:30:00

# Obtendo a data e o horário separadamente
print(dt.date())  # 2022-02-10
print(dt.time())  # 14:30:00

# Criando um intervalo de tempo
delta = datetime.timedelta(hours=1, minutes=30)
print(delta)  # 1:30:00

# Adicionando/subtraindo um intervalo de tempo a uma data/hora
new_dt = dt + delta
print(new_dt)  # 2022-02-10 16:00:00

# Obtendo a hora atual em um determinado fuso horário
tz = datetime.timezone(datetime.timedelta(hours=-3))
now = datetime.datetime.now(tz)
print(now)  # 2022-02-10 13:30:00.063742-03:00
```
