## **CONTROLE DE VENDAS DO DIA**

*Contexto*

Você está desenvolvendo um sistema simples para registrar as vendas do dia.
O sistema deve permitir o usuário inserir o valor de vendas até decidir encerrar.

**Regras do Exercício**

1. Crie uma lista vazia para armazenar os valores das vendas.
2. Use o While para continuar pedindo valores ao usuário.
3. O usuário pode digitar:
- Um número -> representa uma venda.
- sair -> encerra o programa.
4. Regras adicionais:
- Não aceitar valores negativos.
- Se o usuário digitar algo inválido, pedir novamente.

```python
vendas = []

while True:
    entrada = input("Digite o valor da venda (ou 'sair' para encerrar): ").strip().lower()

    #condição de saída
    if entrada  == "sair":
        break

    try:
        valor = float(entrada)

        #validacao de valor negativo
        if valor <= 0:
            print("Valor inválido! Não pode ser negativo")
            continue

        vendas.append(valor)

    except ValueError:
        print("Entrada invalida! Digite o número ou 'sair'!")

#Resultados

if len(vendas) == 0:
    print("\nNenhuma venda registrada")
else:
    total = sum(vendas)
    quantidade = len(vendas)
    maior = max(vendas)
    menor = min(vendas)
    media = total / quantidade

    print("Resumo das Vendas")
    print(f"Total vendido: R$ {total:.2f}")
    print(f"Quantidade: {quantidade}")
    print(f"Maior Venda: R$ {maior}")
    print(f"Menor Venda: R$ {menor}")
    print(f"Media das vendas: R$ {media:.2f}")
```

```python
import csv

with open("vendas.csv", "w", newline="", encoding="utf-8") as arquivo:
    writer = csv.writer(arquivo)
    
    writer.writerow(["Venda"])  # cabeçalho
    
    for venda in vendas:
        writer.writerow([venda])
```
