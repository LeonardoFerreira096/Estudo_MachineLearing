### Loops Aninhados

```python
# Loops aninhados
for i in range(0,5):
    for a in range(0,5):
        print(i, a)
```

```python
# Operando os valores de uma lista com loop for
listaB = [32,53,85,10,15,17,19]
soma = 0 

for i in listaB:
    valor = i * 2
    soma = soma + valor
    print(soma, valor)

print(soma)
```

```python
# Loops em lista de listas
listas = [[1,2,3], [10,15,14], [10.1,8.7,2.3]]
for valor in listas:
    print(valor)
    for z in valor:
        print(z)
```

```python
# Contando os itens de uma lista
lista = [5,6,10,13,17]
contador = 0
for item in lista:
    contador = contador + 1
    
    
print(contador)
```

```python
# Contando o número de colunas
lst = [[1,2,3],[3,4,5,9],[5,6,7]]
primeira_linha = lst[1]
count = 0

for c in primeira_linha:
    count = count + 1
    
print(count)
```

```python
# Pesquisando em listas
listaC = [5, 6, 7, 10, 50]

# Loop através da lista
for uva in listaC:
    if uva == 7:
        print("Número encontrado na lista!")
```

### While

```python
# Usando o loop while para imprimir os valores de 0 a 9
contador = 0
while contador < 10:
    print(contador)
    contador = contador + 1
```

```python
# Também é possível usar a claúsula else para encerrar o loop while
banana = 0

while banana < 10:
    print ('O valor de banana nesta iteração é: ', banana)
    print (' banana ainda é menor que 10, somando 1 a banana')
    banana = banana + 1
    
else:
    print ('Loop concluído!')
```

### Pass, Break, Continue

```python
contador = 0
while contador < 100:
    print(contador)
    if contador == 4:
        break
    else:
        pass    
    contador = contador + 1
```

```python
for v in "Python":
    if v == "h" or v == "n":
        continue
    print(v)
```

### While e For juntos

```python
for i in range(1,300):
    j = 2
    contador = 0
    while j < i:
        if i % j == 0:
            contador = 1
            j = j + 1
        else:
            j = j + 1
    
    if contador == 0:
        print(str(i) + " é um número primo")
        contador = 0
    else:
        contador = 0
```
