```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criando um conjunto de dados simples (tipo uma tabela de banco de dados)
data = {
    'Mês': ['Jan', 'Fev', 'Mar'],
    'Vendas': [1500, 2200, 1800]
}

df = pd.DataFrame(data)

# Exibindo a tabela
print("Tabela de Vendas:")
print(df)

# Criando um gráfico rápido
df.plot(x='Mês', y='Vendas', kind='bar', color='skyblue')
plt.title('Vendas por Mês')
plt.show()
```
